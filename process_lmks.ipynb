```python
import xml.etree.ElementTree as Et
from xml.etree.ElementTree import Element, ElementTree

import numpy as np
import cv2, os
import pandas as pd

from matplotlib import pyplot as plt


img_size = 224

def resize_img(im):
    old_size = im.shape[:2] # old_size is in (height, width) format
    ratio = float(img_size) / max(old_size)
    new_size = tuple([int(x*ratio) for x in old_size])
    # new_size should be in (width, height) format
    im = cv2.resize(im, (new_size[1], new_size[0]))
    delta_w = img_size - new_size[1]
    delta_h = img_size - new_size[0]
    top, bottom = delta_h // 2, delta_h - (delta_h // 2)
    left, right = delta_w // 2, delta_w - (delta_w // 2)
    new_im = cv2.copyMakeBorder(im, top, bottom, left, right, cv2.BORDER_CONSTANT,
        value=[0, 0, 0])
    return new_im, ratio, top, left
```

```python
from matplotlib import pyplot as plt


 
for directory in ['rot_york_train', 'rot_york_validation']:
    base_path = "D:/Developer/2021___1/AI/Datasets/york/{}".format(directory)
    file_list = sorted(os.listdir(base_path))
    
    dataset = {
        'imgs': [],
        'lmks': []
    }
    
    print("size : ", len(file_list))
    
    for idx, f in enumerate(file_list):
      if '.cat' not in f and '.dog' not in f:
          continue
    
      # if idx < 200 or idx > 380 :
      #   print("idx : ", idx)
      #   continue
    
      img_filename, ext = os.path.splitext(f)
      print("image name = " + str(idx) + " | " + img_filename)
      origin_filename, origin_ext = os.path.splitext(img_filename)
      # print("origin_filename name = " + origin_filename)
    
      pd_frame = pd.read_csv(os.path.join(base_path, f), sep=' ', header=None)
      landmarks = (pd_frame.values[0][1:-1]).reshape((-1, 2))
    
      try:
        tree = Et.parse(os.path.join(base_path, origin_filename + ".xml"))
      except:
        continue
    
      root = tree.getroot()
    
      objects = root.findall("object")
      bb = []
    
      if len(objects) > 1:
        continue
    
    
      
      img = cv2.imread(os.path.join(base_path, img_filename))
    
      # try:
      #   img, ratio, top, left = resize_img(img)
      # except Exception as ex:
      #   print("Exception : ", ex)
      #   continue
    
      for _object in objects:
        name = _object.find("name").text
        bndbox = _object.find("bndbox")
        xmin = bndbox.find("xmin").text
        ymin = bndbox.find("ymin").text
        xmax = bndbox.find("xmax").text
        ymax = bndbox.find("ymax").text
    
        
        bb.append(int(float(xmin)))
        bb.append(int(float(ymin)))
        bb.append(int(float(xmax)))
        bb.append(int(float(ymax)))
        break
    
      bb = np.array([[int(float(xmin)), int(float(ymin))],[int(float(xmax)), int(float(ymax))]])
      # bb = ((bb * ratio) + np.array([left, top])).astype(np.int)
      bb = np.clip(bb, 0, 99999)
      new_landmarks = landmarks - bb[0]
    
      try:
        new_img = img[bb[0][1]:bb[1][1], bb[0][0]:bb[1][0]]
      except:
        continue
      
      try:
        img, ratio, top, left = resize_img(new_img)
      except:
        continue
    
      new_landmarks = ((new_landmarks * ratio) + np.array([left, top])).astype(np.int) # padding 붙여준것에 대한 값 조정
    
      dataset['imgs'].append(img)
      dataset['lmks'].append(new_landmarks[0:3].flatten())
    
    
        
    np.save('D:/Developer/2021___1/AI/Datasets/york/{}.npy'.format(directory), np.array(dataset))    
        
    print("dataset len : ", len(dataset.get('imgs')))
      
 
 
print("XML parsing END")
```
