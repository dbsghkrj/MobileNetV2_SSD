```python
import os
import glob
import pandas as pd
import xml.etree.ElementTree as ET


def xml_to_csv(path):
	xml_list = []
	for xml_file in glob.glob(path + '/*.xml'):
		tree = ET.parse(xml_file)
		root = tree.getroot()
		base=os.path.basename(xml_file)
		print(os.path.splitext(base)[0])
		for member in root.findall('object'):
				value = (os.path.splitext(base)[0]+'.jpg',
									int(root.find('size')[0].text),
									int(root.find('size')[1].text),
									member[0].text,
									int(member[4][0].text),
									int(member[4][1].text),
									int(member[4][2].text),
									int(member[4][3].text)
									)
				xml_list.append(value)
	column_name = ['filename', 'width', 'height', 'class', 'xmin', 'ymin', 'xmax', 'ymax']
	xml_df = pd.DataFrame(xml_list, columns=column_name)
	return xml_df

# for directory in ['poodle_train', 'poodle_validation', 'rot_pome_train', 'rot_pome_validation']:
for directory in ['rot_inc_pome_train_600', 'rot_inc_pome_validation_600']:
		image_path = os.path.join(os.getcwd(), 'D:/Developer/2021___1/AI/Datasets/pomeranian/{}'.format(directory).format(directory))
		xml_df = xml_to_csv(image_path)
		xml_df.to_csv('D:/Developer/2021___1/AI/Datasets/pomeranian/{}_labels.csv'.format(directory), index=None)
		print('Successfully converted xml to csv.')
```
